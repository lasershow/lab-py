```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# データ準備

```python
tips = sns.load_dataset("tips")
```

## jointplot

```python
g = sns.jointplot(tips.columns[0], tips.columns[1], tips)
```

## pairplot

```python
sns.pairplot(tips,hue="sex")
```

- hueには、層となる情報を入れる（数値以外の情報をよく使っているように見える。）

```python
sns.pairplot(tips,hue="sex", kind='reg', diag_kind='kde')
```

## facetgrid

```python
train_df = pd.read_csv('testdata/train.csv')
test_df = pd.read_csv('testdata/test.csv')
```

### hist
mapは層だけ。

```python
g = sns.FacetGrid(train_df, col='Survived')
g.map(plt.hist, 'Age', bins=20)
```

```python
# grid = sns.FacetGrid(train_df, col='Pclass', hue='Survived')
grid = sns.FacetGrid(train_df, col='Survived', row='Pclass', size=2.2, aspect=1.6)
grid.map(plt.hist, 'Age', alpha=.5, bins=20)
grid.add_legend();
```

### pointplot
横軸：mapの1番目の要素
縦軸：mapの2番目の要素
層：mapの3番目の要素

```python
# grid = sns.FacetGrid(train_df, col='Embarked')
grid = sns.FacetGrid(train_df, row='Embarked', size=2.2, aspect=1.6)
grid.map(sns.pointplot, 'Pclass', 'Survived', 'Sex', palette='deep')
grid.add_legend()
# Pclass: 1,2,3
# Survived: 1,2
# Sex: male, female
```
