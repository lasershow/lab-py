## how to paint

### 1 firgure

```python
x = np.linspace(0, 6, 20)
y1 = x
y2 = x ** 2

# figure は 1 つ
plt.figure(figsize=(3, 4))

plt.subplot(2,1,1)
plt.plot(x, y1)

plt.subplot(2,1,2)
plt.plot(x, y2)

plt.show()
```

### 2 figures

```python
x = np.linspace(0, 6, 20)
y1 = x
y2 = x ** 2

# figure 1 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y1)
plt.plot(x, y2)

# figure 2 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y2)
plt.plot(x, y1)

plt.show()
```

```python
x = np.linspace(0, 6, 20)
y1 = x
y2 = x ** 2

# figure 1 つ目
plt.figure()
plt.plot(x, y1)

# figure 2 つ目
plt.figure()
plt.plot(x, y2)

plt.show()
```

### what is subplot

```python
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4

plt.figure(figsize=(6,4))

# 左上
plt.subplot(2,2,1)
plt.plot(x, y1)

# 右上
plt.subplot(2,2,2)
plt.plot(x, y2)

# 左下
plt.subplot(2,2,3)
plt.plot(x, y3)

# 右下
plt.subplot(2,2,4)
plt.plot(x, y4)

plt.show()
```

### line details

```python
plt.plot(x, np.sin(x - 0), color='blue')        # 色の名前
plt.plot(x, np.sin(x - 1), color='g')           # 色の頭文字
plt.plot(x, np.sin(x - 2), color='0.75')        # グレースケール
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB
```

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
plt.plot(x, x + 0, '-g')  # solid green
plt.plot(x, x + 1, '--c') # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r');  # dotted red
```

# 演習

```python
 x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
 
fig = plt.figure(figsize=(6, 4))
 
# 左上
ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, y1)
 
# 右上
ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, y2)
 
# 左下
ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, y3)
 
# 右下
ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, y4)
```

```python
x = np.linspace(0, 10, 20)
y1 = x
y2 = x ** 2 


# figure は 1 つ
plt.figure(figsize=(3, 4))

plt.subplot(2,1,1)
plt.plot(x, y1)

plt.subplot(2,1,2)
plt.plot(x, y2)
```

```python
x = np.linspace(0, 10, 20)
y1 = x
y2 = x ** 2

# figure 1 つ目
plt.figure(figsize=(3, 2))
plt.plot(x, y1)
plt.plot(x, y2)

# figure 2 つ目
plt.figure(figsize=(3, 2))


plt.plot(x, y1)
```

```python
x = np.linspace(-3, 3, 20)
y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4

fig, ax = plt.subplots(2, 2, figsize=(6, 4))

# 左上
ax[0, 0].plot(x, y1)

# 右上
ax[0, 1].plot(x, y2)

# 左下
ax[1, 0].plot(x, y3)

# 右下
ax[1, 1].plot(x, y4)
```
